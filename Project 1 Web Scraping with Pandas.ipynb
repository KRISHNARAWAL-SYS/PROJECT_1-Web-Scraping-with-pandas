# 2.1 Multiple Leagues

##### https://www.football-data.co.uk/mmz4281/2122/E3.csv
##### https://www.football-data.co.uk/mmz4281/2122/EC.csv

```python
# creating a list of leagues
leagues = ['E0' , 'E1' , 'E2' , 'E3' , 'EC']
frames = []

# looping through leagues, read multiple CSVs and append it into list
for league in leagues:
    df = pd.read_csv( root + '/2122' + '/' + league + '.csv' , encoding = 'unicode_escape')
    frames.append(df)
```

```python
# creating a list of leagues
leagues = ['E0' , 'E2' , 'E3' ]
frames = []

# looping through leagues, read multiple CSVs and append them to a list
for league in leagues:
    for season in range(15 , 21):
        df = pd.read_csv( root + str(season) + str(season+1)  + '/' + league + '.csv' , encoding = 'unicode_escape')
        df.insert(1 , 'season' , season )
        frames.append(df)

# deleted few leagues to make data less crowded
```

# 3 Organize all the data in a Dictionary

```python
# create a dictionary with original name of league as key
dict_countries = {
    'Spanish La Liga':'SP1' , 'Spanish Segunda Division' : 'SP2','German Bundesliga':'D1',
    'English Premier League':'E0' , 'English League 1 ': 'E1' , 'English League 2' : 'E2'
}
                 
```

```python
# get a dictionary element
dict_countries['Spanish La Liga']
```

```python
for league in dict_countries:
    print(dict_countries[league])    # ---> to print value
#   print(league)    --> to print key
```

```python
dict_historical_data={}

#creating list of leagues
# leagues = ['E0' , 'E01' , 'E02']


# looping through leagues, read multiple CSVs and append them to a list
for league in dict_countries:
    frames = [] 
    for season in range(15 , 21):
        df = pd.read_csv( root + str(season) + str(season+1)  + '/' + dict_countries[league] + '.csv' , encoding = 'unicode_escape')
        df.insert(1 , 'season' , season )
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league]= df_concat
```
